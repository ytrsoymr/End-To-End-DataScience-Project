## 6.  Univariate Analysis

- Distributions of numerical features (histograms, KDE plots)
- Frequency of categorical values (bar plots)
- Outlier detection using boxplots


```python
sns.histplot(x=iris["petal_length"])
```

```python
import matplotlib.pyplot as plt

iris.hist(figsize=(10, 8), color='skyblue', edgecolor='black')
plt.suptitle("Univariate Histograms of Iris Features", fontsize=16)
plt.show()
```

```python
import seaborn as sns

for col in iris.columns[:-1]:
    sns.boxplot(data=iris, x='species', y=col)
    plt.title(f"Box Plot of {col} by Species")
    plt.show()
```

```python
for col in iris.columns[:-1]:
    sns.kdeplot(data=iris, x=col, hue='species', fill=True)
    plt.title(f"Density Plot of {col}")
    plt.show()
```

```python
for col in iris.columns[:-1]:
    sns.violinplot(data=iris, x='species', y=col)
    plt.title(f"Violin Plot of {col} by Species")
    plt.show()
```

## 7. 🔗 Bivariate Analysis

- Correlation matrix and heatmap
- Scatter plots for numerical vs numerical
- Boxplots/violin plots for categorical vs numerical
- Crosstabs for categorical vs categorical


```python
iris.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Width by Species')
plt.show()

sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Width by Species')
plt.show()
```

```python
sns.pairplot(iris, hue='species', diag_kind='kde', corner=True)
```

```python
corr = iris.drop(columns='species').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```
